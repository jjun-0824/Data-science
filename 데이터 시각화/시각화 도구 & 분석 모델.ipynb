```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()

# 선형식 그리기
xsample=[1,2,3,4,5]
ysample=[6,7,8,9,10]
plt.xlabel("")
plt.ylabel("")
plt(xsample,ysample)
plt.show()

# 개수 히스토그램 그리기(데이터의 분포도를 파악)
sns.catplot(data=df,x="label",kind="count")

# x와 y변수의 다차원 분포 파악하기(hue는 보통 분류값들을 지정)
sns.catplot(data=df, x="SepalLengthCm", y="SepalWidthCm",hue="Species",aspect=2)

# 데이터의 밀집도 파악(데이터가 고르게 분포되어있는지? 데이터의 수가 편향되어 있으면 특정 값들에 대해서만 학습이 잘되는 경우가 존재)
facet=sns.FacetGrid(df,hue="Species",aspect=2)
facet.map(sns.kdeplot,"SepalLengthCm")
facet.add_legend()
plt.show()

# boxplot 그리기
sns.catplot(kind="box", data=df,x="Species",y="SepalLengthCm",aspect=2)
sns.violinplot(data=df,x="Species",y="PetalLengthCm")

#violinplot: 값의 범위와 밀집도를 한꺼번에 확인할 수 있다.
#일반 catplot을 사용할 시에는 구분된다는 것은 알 수 있지만 유효범위를 명확히 알 수 없음
# boxplot은 가능 but 밀집도는 알수 없음=>kedplot 참고
# kedplot+boxplot=>violin plot


facet=sns.FacetGrid(df,aspect=4)
facet.map(sns.kdeplot,"quality")
facet.add_legend()
plt.show()

# 변수들간의 상관계수 구하기
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(),annot=True,fmt=".2f",square=True)
plt.show()
```
