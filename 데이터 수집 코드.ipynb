```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
# from selenium.webdriver.common.action_chains import ActionChains
import time

item_name = []

# 크롬창(웹드라이버) 열기
driver = webdriver.Chrome("./chromedriver")
# ac = ActionChains(driver)

# ootd 태그 검색결과 페이지 접속
driver.get("https://www.glowpick.com/beauty/ranking?id=7&level=2&order=rank")

# 스크롤 맨 하단까지 내리기
search_bt = driver.find_element_by_css_selector('button.gp-header__search-toggle-btn')
for i in range(23):
    search_bt.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)

# 제품명 수집하기
item = driver.find_elements_by_css_selector("div.list-item")
for n in item:
    name = n.find_element_by_css_selector("p.details__labels__name").text
    item_name.append(name)
    # print(name)

# 제품 개별 검색하기
for i in range(0,99):

     search_bt.click()
     search_box = driver.find_element_by_css_selector('input.search-box__input')
     search_box.send_keys(item_name[i])
     time.sleep(1)

     search_box.send_keys(Keys.ENTER)

     time.sleep(2.5)

     result_box = driver.find_element_by_css_selector('div.product-image-wrapper')
     result_box.click()

     time.sleep(6)

##
     stores1 = driver.find_elements_by_css_selector("span.product-main-info__product_name__text")
     print(stores1[0].text)
     stores2 = driver.find_elements_by_css_selector("span.product-main-info__volume_price--bold")
     print(stores2[0].text)
     stores3 = driver.find_elements_by_css_selector("span.info__tags")
     for m in range(len(stores3)):
         print(stores3[m].text)

##

     driver.execute_script("window.history.go(-1)")
     time.sleep(1)


     i += 1
```
