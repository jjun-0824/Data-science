```python
import pandas as pd
```

```python
df=pd.read_csv("2019_view_rating.csv")
```

```python
df
```

```python
df=df.drop(1440,axis=0)
```

```python
df1=df[["시간대","2019-01-01 to 2019-12-31"]]
```

```python
df1
```

```python
rate=[]
for j in range(48):
    total=0
    for i in range(30):
        total=total+df1["2019-01-01 to 2019-12-31"][30*j+i]
    rate.append(total)
rate
```

```python
time=[]
for j in range(48):
    for i in range(30):
        if i==0:
            time.append(df1["시간대"][30*j+i])
time
```

```python
df2=pd.DataFrame({
    "시간대":time,
    "시청률":rate
})
df2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.catplot(data=df2, x="시간대", y="시청률",aspect=5)
```

```python
# 시청률 0.15 넘는 시간대를 프라임 타임이라고 볼 것
df3=df2.loc[df2["시청률"]>0.15,"시간대"]
```

```python
df3=pd.DataFrame({
    "prime":df3
})
df3
# 즉 시간대가 10시, 11시, 13시, 14시, 15시, 16시, 17시, 18시, 23시, 00시 일때 프라임 시간대 1
```

```python
time=[10,11,13,14,15,16,17,18,23,"00"]
df3=pd.DataFrame({
    "prime":time
})
df3
```

```python
# 이 시간대에 포함될 시에 가중치 1두기
df=pd.read_csv("기온강수량.csv")
```

```python
df
```

```python
arr=[]

for i in df.index:
    cnt=0
    for j in df3["prime"]:
        time=str(df["방송일시"][i])[11:13]
        if str(j) in time:
            arr.append(1)
            break
        else:
            cnt=cnt+1
    if cnt==10:
        arr.append(0)
```

```python
df["프라임시간대"]=arr
df
```

```python
df.to_csv("실적데이터_전처리_최종.csv",index=False)
```
