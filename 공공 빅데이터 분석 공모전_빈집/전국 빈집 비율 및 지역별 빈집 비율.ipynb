```python
df=pd.read_csv("빈집의_종류_및_건축연도별_주택시군구_20201026205459.csv")
```

```python
df
```

```python
# 전국 빈집 현황(2015-2019)
data1=df.loc[0]
data1=data1[["2019년","2018년","2017년","2016년","2015년"]]
data1
```

```python
# 시리즈에서 데이터 프레임 변환
df1=pd.DataFrame({
    "년도":["2015년","2016년","2017년","2018년","2019년"],
    "빈집 수": [35635,36515,43218,53926,73121]
})
df1
```

```python
# data1 히스토그램 그리기
sns.set(font_scale=1.2)
sequential_colors = sns.color_palette("Greens", 5)
# sns.palplot(sequential_colors)
sns.set_palette(sequential_colors)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(9,5))
sns.barplot(data=df1,x="년도",y="빈집 수")
```

```python
# 지역별 빈집 현황(2019 기준/단독 주택)
# data2=df.loc[df["주택의 종류별"]=="계"] 추리고
# seaborn을 활용하여 시각화하기
data2=df.loc[df["주택의 종류별"]=="계"]
data2=df.loc[df["주택의 종류별"]=="계"][1:]
data2
```

```python
data2_sort = data2.sort_values(by='주택_계' ,ascending=False)
```

```python
sequential_colors = sns.color_palette("GnBu", 17)
# sns.palplot(sequential_colors)
sns.set_palette(sequential_colors)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize=(30,5))
sns.barplot(data=data2_sort,x="행정구역별(시군구)",y="주택_계")
```
