빈집 가중치 선정 방법:
https://m.blog.naver.com/PostView.nhn?blogId=n96320&logNo=70120582531&proxyReferer=https:%2F%2Fwww.google.com%2F

```python
# 빈 열 만들기
df["년도 점수"]=pd.Series()
df["면적 점수"]=pd.Series()
df["거리 점수"]=pd.Series()

# 철거 가능성/활용가능성 ->설립년도 2020년-2005년도
# 설립년도가 오래되었을수록 노후 가능성이 높다는 전제. 
# 설립년도가 최근일수록 높은 점수 부여
df.loc[df["발생 연도"]<1950,"년도 점수"]=0
df.loc[(df["발생 연도"]>=1950)&(df["발생 연도"]<1975),"년도 점수"]=1
df.loc[(df["발생 연도"]>=1975)&(df["발생 연도"]<2000),"년도 점수"]=2
df.loc[df["발생 연도"]>=2000,"년도 점수"]=3

# 면적 230 이하: 1인 가구 최소 주거 면적 참고
# 1인 가구: 0-26 -->0
# 2-3인 가구(적정 수준, 만약 가구당 거주인구를 1-2인으로 한정할 시): 26-43 --> 2점
# 4인 가구 이상(43-): 면적이 넓어짐에 따른 리모델링 비용 증가 고려, 다만 가족 단위 거주일 경우에 필요할 수 있으므로 1점 -->1
적정 면적은 현행법상 최소 주거 면적을 참고하여 26-43M^2로 선정.
평수가 작을 경우에는 생활의 불편함을, 평수가 지나치게 클 경우에는 리모델링 비용의 과도한 증가를 초래하기 때문.

df.loc[df["연면적(㎡)"]<26,"면적 점수"]=0
df.loc[(df["연면적(㎡)"]>=26)&(df["연면적(㎡)"]<43),"면적 점수"]=2
df.loc[df["연면적(㎡)"]>=43,"면적 점수"]=1
df.loc[df["연면적(㎡)"]>=60,"면적 점수"]=0

# 시내와의 거리: 생활 서비스 접근성 및 인프라 관련,
# 관광지와의 거리 0인것은 제외. 정보가 없는 경우
df=df.loc[(df["시내와의 거리"]<=15)&(df["시내와의 거리"]>0)]
# 시내와의 거리를 가까울수록 높은 가산점
# 0-10: 3, 10-15: 2, 15-20:1  20-:0
df.loc[df["시내와의 거리"]<10,"거리 점수"]=3
df.loc[(df["시내와의 거리"]>=10)&(df["시내와의 거리"]<15),"거리 점수"]=2
df.loc[(df["시내와의 거리"]>=15)&(df["시내와의 거리"]<20),"거리 점수"]=1
df.loc[df["시내와의 거리"]>=20,"거리 점수"]=0

```

# AHP 기법을 활용한 가중치 선정
년도 점수: 0.347
면적 점수: 0.198
거리 점수: 0.455

```python
# 각각의 변수에 대한 가산점 부여
df["종합 점수"]=pd.Series()
for i in df.index:
    df["종합 점수"][i]=(0.347)*df["년도 점수"][i]+(0.198)*df["면적 점수"][i]+(0.455)*df["거리 점수"][i]
```
