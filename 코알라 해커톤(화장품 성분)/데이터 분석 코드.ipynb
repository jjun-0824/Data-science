```python

# 데이터 불러오기
import pandas as pd
import numpy as np


df = pd.read_csv('glowpick_data.csv', thousands = ',')
df.head(10)
```

```python

df.describe()
def remove_ml(x):
     return x.replace('ml', '')
def remove_g(x):
     return x.replace('g', '')
     
#g, ml 지우고 형변환

df['volume'] = df['volume'].apply(remove_ml)
df['volume'] = df['volume'].apply(remove_g)
df.loc[ df['volume'] == '15*2',"volume"] = 30
df.loc[ df['volume'] == '14*2',"volume"] = 28
df.loc[ df['volume'] == '13*2',"volume"] = 26
df.loc[ df['volume'] == '12.5*2',"volume"] = 25
df.loc[ df['volume'] == '12*2',"volume"] = 24
df.loc[ df['volume'] == '13*2ea',"volume"] = 26
df.loc[ df['volume'] == '0.3*8',"volume"] = 2.4
df['volume'] = pd.to_numeric(df['volume'])
df['price'] = pd.to_numeric(df['price'])


df.head(20)
```

```python
#용량당 가격 열 만들기

df['price/volume'] = df['price'] / df['volume']

df.head()
```

```python
#랭킹 열 만들기
num = list(range(1, 101))
num
df['rank'] = num
df.head()
```

```python
#필요없는 열 지우기
df = df.drop(['name','brand','volume','price','tag'], axis=1)

df.head()
```

```python

input_data = df.drop('rank', axis=1)
target_data = df['rank']

input_data
target_data
print(input_data.shape, target_data.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input_data, target_data)
```

```python

#모델링 학습: Random Forest
    
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators=100)

forest.fit(x_train, y_train)
```

```python

print('train set accuracy:', forest.score(x_train, y_train))
```

```python
# 모델검증

print('test set accuracy:', forest.score(x_test,y_test))
```

```python
y_predict = forest.predict(x_test)
comparison = pd.DataFrame(y_test)
comparison['내 추측'] = y_predict

comparison
```
